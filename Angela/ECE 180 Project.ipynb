```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fname1 = "Educational attainment and labour-force status.csv"

df1 = pd.read_csv(fname1)
df1 = df1[["Country","ISCED 2011 A education level","Gender","Age","Measure","Value"]]
df1 = df1[df1.Gender == "Total"]
df1 = df1[df1.Measure == "Value"]
df1 = df1[df1["ISCED 2011 A education level"] == "Tertiary education"]
df1 = df1[df1.Country != "OECD - Average"]
df1 = df1[df1.Country != "European Union 22 members in OECD"]
df1 = df1[df1.Country != "G20"]
df1 = df1.drop(["ISCED 2011 A education level","Gender", "Age", "Measure"], 1)
df1 = df1.set_index("Country")
df1 = df1.loc[["Australia","Brazil","Canada","France","Korea","Mexico","Sweden","United Kingdom","United States"]]
df1.columns = ["Tertiary Attainment (%)"]
df1
```

```python
fname2 = "Percentage in education not in education.csv"

df2 = pd.read_csv(fname2)
df2 = df2[["Country","ISCED-A 2011","Gender","EDU_LFS_STATUS","Education and labour force status","Measure","Value"]]
df2 = df2[df2.Gender == "Total"]
df2 = df2[df2.Measure == "Value"]
df2 = df2[df2["ISCED-A 2011"] == "Tertiary education"]
df2 = df2[df2["EDU_LFS_STATUS"] != "ED_E"]
df2 = df2[df2["EDU_LFS_STATUS"] != "ED_E_WS"]
df2 = df2[df2["EDU_LFS_STATUS"] != "ED_E_OTH"]
df2 = df2[df2["EDU_LFS_STATUS"] != "ED_U_I"]
df2 = df2[df2["EDU_LFS_STATUS"] != "ED_I"]
df2 = df2[df2["EDU_LFS_STATUS"] != "ED_U"]
df2 = df2[df2["EDU_LFS_STATUS"] != "NE"]
df2 = df2[df2["EDU_LFS_STATUS"] != "NE_I"]
df2 = df2[df2["EDU_LFS_STATUS"] != "NE_U"]
df2 = df2[df2.Country != "OECD - Average"]
df2 = df2[df2.Country != "European Union 22 members in OECD"]
df2 = df2[df2.Country != "G20"]
df2 = df2.drop(["ISCED-A 2011","Gender","EDU_LFS_STATUS","Measure"], 1)
df2
```

```python
fname3 = "Public expenditure on education as percentage of total government expenditure.csv"

df3 = pd.read_csv(fname3)
df3 = df3[["Country","Education level and programe orientation","Institution type","Value"]]
df3 = df3[df3["Institution type"] == "All sectors"]
df3 = df3[df3["Education level and programe orientation"] == "Total tertiary education (ISCED2011 levels 5 to 8)"]
df3 = df3[df3.Country != "OECD - Average"]
df3 = df3[df3.Country != "European Union 22 members in OECD"]
df3 = df3[df3.Country != "G20"]
df3 = df3.drop(["Education level and programe orientation","Institution type"], 1)
df3 = df3.set_index("Country")
df3 = df3.loc[["Australia","Brazil","Canada","France","Korea","Mexico","Sweden","United Kingdom","United States"]]
df3.columns = ["Public Expenditure (%)"]
df3
```

```python
# Add Tertiary Attainment column from df1 to df3
df4 = df3.join(df1)
df4
```

```python
ax = df4.plot.scatter(x = "Public Expenditure (%)", y = "Tertiary Attainment (%)")
# Label each data point with country
for k,v in df4.iterrows():
    ax.annotate(k,v, xytext=(5,-5), textcoords="offset points", fontsize=8)
plt.title("Tertiary Attainment Based on Percentage of Public Expenditure\n")
plt.show()
```
